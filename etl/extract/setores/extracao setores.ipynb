```python
import pandas as pd;
import yfinance as yf;
```

```python
def get_all_b3_tickers_info():
    b3_tickers_list = list(pd.read_csv('acoes-listadas-b3.csv').Ticker);
    b3_tickers_list = [ticker + '.SA' for ticker in b3_tickers_list];
    start_date = '2008-01-01';
    b3_tickers = yf.Tickers(b3_tickers_list);
    return b3_tickers;
```

```python
def get_sectors(b3_tickers):
    sectors = [];
    tickers = list(b3_tickers.tickers.keys());
    for ticker, pos in zip(tickers, range(0, len(tickers))):
        print(f'{pos}/{len(tickers)}');
        sectors.append(b3_tickers.tickers[ticker].info['sector'] if 'sector' in b3_tickers.tickers[ticker].info else None);
    return tickers, sectors;
```

```python
def save(tickers, sectors):
    series = pd.Series(index=tickers, data=sectors);
    series.to_csv('sectors.csv');
```

```python
def main():
    b3_tickers = get_all_b3_tickers_info();
    tickers, sectors = get_sectors(b3_tickers);
    save(tickers, sectors);
```

```python
main();
```
